```python
import pandas as pd

datata=pd.read_csv('homework1.csv')

# print(df.to_string())

datata=datata.drop(0)
datata=datata.iloc[:,[2,4,7,8,11,12,15,16,19,20,23,24,27,28,31,32,33,34,35,37,38,39,41,42,45,46,47,49,50]]
datata
```

```python
list=[]
for i in datata:
    list.append(i)
list.remove('site_id')
```

```python
datata['ungraduated_sum']=0
for i in list:
    datata[i]=datata[i].astype(int)
    
list.remove('edu_age_15up_total')
```

```python
for j in list:
    datata['ungraduated_sum']+=datata[j]
datata=datata.groupby(['site_id'])['ungraduated_sum','edu_age_15up_total', 'edu_doctor_ungraduated_m', 'edu_doctor_ungraduated_f', 'edu_master_ungraduated_m', 'edu_master_ungraduated_f', 'edu_university_ungraduated_m', 'edu_university_ungraduated_f', 'edu_juniorcollege_2ys_ungraduated_m', 'edu_juniorcollege_2ys_ungraduated_f', 'edu_juniorcollege_5ys_final2y_ungraduated_m', 'edu_juniorcollege_5ys_final2y_ungraduated_f', 'edu_senior_ungraduated_m', 'edu_senior_ungraduated_f', 'edu_seniorvocational_ungraduated_m', 'edu_seniorvocational_ungraduated_f', 'edu_juniorcollege_5ys_first3y_ungraduated_m', 'edu_juniorcollege_5ys_first3y_ungraduated_f', 'edu_junior_graduated_m', 'edu_junior_ungraduated_m', 'edu_junior_ungraduated_f', 'edu_juniorvocational_graduated_m', 'edu_juniorvocational_ungraduated_m', 'edu_juniorvocational_ungraduated_f', 'edu_primary_ungraduated_m', 'edu_primary_ungraduated_f', 'edu_selftaughtl_m', 'edu_illiterate_m', 'edu_illiterate_f'].sum().reset_index()
datata
```

```python
datata['rate']=datata['ungraduated_sum']/datata['edu_age_15up_total']
datata["rate_percentage"] = datata["rate"].apply(lambda x: format(x,".2%"))
datata
```

```python
datata['ungraduated_sum_rank'] = datata.rate.rank(method='first',ascending=True)

datata=datata.groupby(['site_id'])['ungraduated_sum','rate',"rate_percentage","ungraduated_sum_rank",'edu_age_15up_total', 'edu_doctor_ungraduated_m', 'edu_doctor_ungraduated_f', 'edu_master_ungraduated_m', 'edu_master_ungraduated_f', 'edu_university_ungraduated_m', 'edu_university_ungraduated_f', 'edu_juniorcollege_2ys_ungraduated_m', 'edu_juniorcollege_2ys_ungraduated_f', 'edu_juniorcollege_5ys_final2y_ungraduated_m', 'edu_juniorcollege_5ys_final2y_ungraduated_f', 'edu_senior_ungraduated_m', 'edu_senior_ungraduated_f', 'edu_seniorvocational_ungraduated_m', 'edu_seniorvocational_ungraduated_f', 'edu_juniorcollege_5ys_first3y_ungraduated_m', 'edu_juniorcollege_5ys_first3y_ungraduated_f', 'edu_junior_graduated_m', 'edu_junior_ungraduated_m', 'edu_junior_ungraduated_f', 'edu_juniorvocational_graduated_m', 'edu_juniorvocational_ungraduated_m', 'edu_juniorvocational_ungraduated_f', 'edu_primary_ungraduated_m', 'edu_primary_ungraduated_f', 'edu_selftaughtl_m', 'edu_illiterate_m', 'edu_illiterate_f'].sum().reset_index()

datata.sort_values("ungraduated_sum_rank",inplace=True)

datata
```
