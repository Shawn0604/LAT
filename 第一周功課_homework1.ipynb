```python
import pandas as pd

df=pd.read_csv('week2_homework.csv')

# print(df.to_string())

df
```

```python
df=df.drop(0)
df=df.iloc[:,[2,4,7,8,11,12,15,16,19,20,23,24,27,28,31,32,33,34,35,37,38,39,41,42,45,46,47,49,50]]
df
```

```python
list=[]
for i in df:
    list.append(i)
list.remove('site_id')
print(list)
df['sum']=0
for i in list:
    df[i]=df[i].astype(int)
list.remove('edu_age_15up_total')
for j in list:
    df['sum']+=df[j]
    
#     print(i)
#     df[i]=df[i].astype(int)
# df=df.astype(int)
df=df.groupby(['site_id'])['sum','edu_age_15up_total', 'edu_doctor_ungraduated_m', 'edu_doctor_ungraduated_f', 'edu_master_ungraduated_m', 'edu_master_ungraduated_f', 'edu_university_ungraduated_m', 'edu_university_ungraduated_f', 'edu_juniorcollege_2ys_ungraduated_m', 'edu_juniorcollege_2ys_ungraduated_f', 'edu_juniorcollege_5ys_final2y_ungraduated_m', 'edu_juniorcollege_5ys_final2y_ungraduated_f', 'edu_senior_ungraduated_m', 'edu_senior_ungraduated_f', 'edu_seniorvocational_ungraduated_m', 'edu_seniorvocational_ungraduated_f', 'edu_juniorcollege_5ys_first3y_ungraduated_m', 'edu_juniorcollege_5ys_first3y_ungraduated_f', 'edu_junior_graduated_m', 'edu_junior_ungraduated_m', 'edu_junior_ungraduated_f', 'edu_juniorvocational_graduated_m', 'edu_juniorvocational_ungraduated_m', 'edu_juniorvocational_ungraduated_f', 'edu_primary_ungraduated_m', 'edu_primary_ungraduated_f', 'edu_selftaughtl_m', 'edu_illiterate_m', 'edu_illiterate_f'].sum().reset_index()
df
```

```python
df['rate']=df['sum']/df['edu_age_15up_total']
df["rate_percentage"] = df["rate"].apply(lambda x: format(x,".2%"))
df
```

```python
df['該地區肄業人數百分比排名'] = df.rate.rank(method='first',ascending=True)

df.sort_values("該地區肄業人數百分比排名",inplace=True)

df
```
